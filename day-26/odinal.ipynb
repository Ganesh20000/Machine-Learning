```python

#! model build 
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.iloc[:,0:2],df.iloc[:,-1],random_state=0,test_size=0.2)
```

```python

#* importing ordinal encoder 
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe= OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
```

```python

#! FIT THE MODEL 
oe.fit(x_train)

#* transform the mode
oe.transform(x_train)
oe.transform(x_test)
```
