```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('train.csv')
```

```python
df
```

```python
df.drop(columns=['Name','Cabin','Ticket','PassengerId'],inplace=True)
```

```python
x_train,x_test,y_train,Y_test=train_test_split(df.drop(columns='Survived'),df['Survived'],test_size=0.2,random_state=0)
```

```python
x_train.shape,x_test.shape
```

```python
df.isnull().sum()
```

#### simple imputer class

```python
s1 = SimpleImputer()
s2= SimpleImputer(strategy='most_frequent')

s1_age_train =s1.fit_transform(x_train[['Age']])
s1_age_test = s1.fit_transform(x_test[['Age']])


emb_train_hot = s2.fit_transform(x_train[['Embarked']])
emb_test_hot  =s2.fit_transform(x_test[['Embarked']])
```

```python
s1_age_train.shape,s1_age_test.shape
```

#### one hot encoding in this

```python
one = OneHotEncoder(handle_unknown='ignore',sparse_output=False)
two = OneHotEncoder(handle_unknown='ignore',sparse_output=False)


sex_train_hot=one.fit_transform(x_train[['Sex']])
emb_train_hot = two.fit_transform(x_train[['Embarked']])


sex_test_hot = one.fit_transform(x_test[['Sex']])
emb_test_hot = two.fit_transform(x_test[['Embarked']])
```

```python
df
```

```python
x_train_rem = x_train.drop(columns=['Sex','Age','Embarked'])
```

```python
x_test_rem = x_test.drop(columns=['Sex','Age','Embarked'])
```

```python
sex_train_hot.shape,emb_train_hot.shape,s1_age_train.shape 
```

```python
x_train_rem.shape,x_test_rem.shape 
```

```python

x_train_transformer =np.concatenate((x_train_rem,sex_train_hot,emb_train_hot,s1_age_train),axis=1)
```

```python
x_train_transformer.shape
```

```python
x_train_transformer
```

```python
x_test_transform= np.concatenate((x_test_rem,s1_age_test,emb_test_hot,sex_test_hot),axis=1)
```

```python
x_test_transform
```

```python
x_test_transform.shape
```

```python
t = DecisionTreeClassifier()
t.fit(x_train_transformer,y_train)
```
