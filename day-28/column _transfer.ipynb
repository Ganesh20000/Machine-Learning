```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
```

```python
s= SimpleImputer()

x_train_fever = s.fit_transform(x_train[['fever']])

x_test_fever =s.fit_transform(x_test[['fever']])
```

```python
x_test_fever.shape,x_train_fever.shape
```

### one ordinal encoding with 

```python
od= OrdinalEncoder(categories=[['Mild','Strong']])

x_train_or = od.fit_transform(x_train[['cough']])

x_test_or= od.fit_transform(x_test[['cough']])
```

```python
x_test_or.shape ,x_train_or.shape
```

#### one hot encoder in this city and gender

```python
one = OneHotEncoder(drop='first')

x_train_one= one.fit_transform(x_train[['gender','city']]).toarray()


x_test_one = one.transform(x_test[['gender','city']]).toarray()
  
```

```python
x_test_one.shape,x_train_one.shape
```
